```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def create_new_list(list1, list2):
    if not list1 or not list2:
        return None

    head = None
    tail = None

    while list1 and list2:
        if list1.data >= list2.data:
            new_node = Node(list1.data)
            list1 = list1.next
        else:
            new_node = Node(list2.data)
            list2 = list2.next

        if not head:
            head = new_node
            tail = new_node
        else:
            tail.next = new_node
            tail = new_node

    while list1:
        new_node = Node(list1.data)
        list1 = list1.next

        if not head:
            head = new_node
            tail = new_node
        else:
            tail.next = new_node
            tail = new_node

    while list2:
        new_node = Node(list2.data)
        list2 = list2.next

        if not head:
            head = new_node
            tail = new_node
        else:
            tail.next = new_node
            tail = new_node

    return head

# Example 1
# Creating linked list 1: 5->2->3->8
n1 = Node(5)
n2 = Node(2)
n3 = Node(3)
n4 = Node(8)
n1.next = n2
n2.next = n3
n3.next = n4
list1 = n1

# Creating linked list 2: 1->7->4->5
n5 = Node(1)
n6 = Node(7)
n7 = Node(4)
n8 = Node(5)
n5.next = n6
n6.next = n7
n7.next = n8
list2 = n5

new_list = create_new_list(list1, list2)

# Printing the new linked list
while new_list:
    print(new_list.data, end=" ")
    new_list = new_list.next
print()

print("---------------")

# Example 2
# Creating linked list 1: 2->8->9->3
n1 = Node(2)
n2 = Node(8)
n3 = Node(9)
n4 = Node(3)
n1.next = n2
n2.next = n3
n3.next = n4
list1 = n1

# Creating linked list 2: 5->3->6->4
n5 = Node(5)
n6 = Node(3)
n7 = Node(6)
n8 = Node(4)
n5.next = n6
n6.next = n7
n7.next = n8
list2 = n5

new_list = create_new_list(list1, list2)

while new_list:
    print(new_list.data, end=" ")
    new_list = new_list.next
print()
```

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def remove_duplicates(head):
    if not head:
        return head

    current = head

    while current.next:
        if current.data == current.next.data:
            current.next = current.next.next
        else:
            current = current.next

    return head

# Example 1
# Creating linked list: 11->11->11->21->43->43->60
n1 = Node(11)
n2 = Node(11)
n3 = Node(11)
n4 = Node(21)
n5 = Node(43)
n6 = Node(43)
n7 = Node(60)
n1.next = n2
n2.next = n3
n3.next = n4
n4.next = n5
n5.next = n6
n6.next = n7
linked_list = n1

new_list = remove_duplicates(linked_list)

while new_list:
    print(new_list.data, end=" ")
    new_list = new_list.next
print()

print("--------------")
# Example 2
# Creating linked list: 10->12->12->25->25->25->34
n1 = Node(10)
n2 = Node(12)
n3 = Node(12)
n4 = Node(25)
n5 = Node(25)
n6 = Node(25)
n7 = Node(34)
n1.next = n2
n2.next = n3
n3.next = n4
n4.next = n5
n5.next = n6
n6.next = n7
linked_list = n1

new_list = remove_duplicates(linked_list)

while new_list:
    print(new_list.data, end=" ")
    new_list = new_list.next
print()
```

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def reverse_k_nodes(head, k):
    if not head:
        return head

    current = head
    next_node = None
    prev = None
    count = 0

    while current and count < k:
        next_node = current.next
        current.next = prev
        prev = current
        current = next_node
        count += 1

    if next_node:
        head.next = reverse_k_nodes(next_node, k)

    return prev

# Example 1
# Creating linked list: 1->2->2->4->5->6->7->8
n1 = Node(1)
n2 = Node(2)
n3 = Node(2)
n4 = Node(4)
n5 = Node(5)
n6 = Node(6)
n7 = Node(7)
n8 = Node(8)
n1.next = n2
n2.next = n3
n3.next = n4
n4.next = n5
n5.next = n6
n6.next = n7
n7.next = n8
linked_list = n1

new_list = reverse_k_nodes(linked_list, 4)

while new_list:
    print(new_list.data, end=" ")
    new_list = new_list.next
print()

print("-----------------")

# Example 2
# Creating linked list: 1->2->3->4->5
n1 = Node(1)
n2 = Node(2)
n3 = Node(3)
n4 = Node(4)
n5 = Node(5)
n1.next = n2
n2.next = n3
n3.next = n4
n4.next = n5
linked_list = n1

new_list = reverse_k_nodes(linked_list, 3)

while new_list:
    print(new_list.data, end=" ")
    new_list = new_list.next
print()
```

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def reverse_alternate_k_nodes(head, k):
    if not head:
        return head

    current = head
    next_node = None
    prev = None
    count = 0
    reverse = True

    while current and count < k:
        next_node = current.next
        if reverse:
            current.next = prev
        prev = current
        current = next_node
        count += 1

    if next_node and count >= k:
        head.next = reverse_alternate_k_nodes(next_node, k)

    return prev

# Example
# Creating linked list: 1->2->3->4->5->6->7->8->9
n1 = Node(1)
n2 = Node(2)
n3 = Node(3)
n4 = Node(4)
n5 = Node(5)
n6 = Node(6)
n7 = Node(7)
n8 = Node(8)
n9 = Node(9)
n1.next = n2
n2.next = n3
n3.next = n4
n4.next = n5
n5.next = n6
n6.next = n7
n7.next = n8
n8.next = n9
linked_list = n1

new_list = reverse_alternate_k_nodes(linked_list, 3)

while new_list:
    print(new_list.data, end=" ")
    new_list = new_list.next
print()
```

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def delete_last_occurrence(head, key):
    if not head:
        return head

    toDelete = None
    prev = None
    current = head

    while current:
        if current.data == key:
            toDelete = prev
        prev = current
        current = current.next

    if toDelete:
        toDelete.next = toDelete.next.next
    else:
        
        if head.data == key:
            head = head.next

    return head

# Example
# Creating linked list: 1->2->3->5->2->10
n1 = Node(1)
n2 = Node(2)
n3 = Node(3)
n4 = Node(5)
n5 = Node(2)
n6 = Node(10)
n1.next = n2
n2.next = n3
n3.next = n4
n4.next = n5
n5.next = n6
linked_list = n1

new_list = delete_last_occurrence(linked_list, 2)

while new_list:
    print(new_list.data, end=" ")
    new_list = new_list.next
print()
```

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

def merge_sorted_lists(a, b):

    dummy = Node()
    tail = dummy


    while a and b:
        if a.data <= b.data:
            tail.next = a
            a = a.next
        else:
            tail.next = b
            b = b.next
        tail = tail.next

    if a:
        tail.next = a
    else:
        tail.next = b

    return dummy.next

# Example 1
# Creating linked lists: a: 5->10->15, b: 2->3->20
a1 = Node(5)
a2 = Node(10)
a3 = Node(15)
a1.next = a2
a2.next = a3
list_a = a1

b1 = Node(2)
b2 = Node(3)
b3 = Node(20)
b1.next = b2
b2.next = b3
list_b = b1

merged_list = merge_sorted_lists(list_a, list_b)

while merged_list:
    print(merged_list.data, end=" ")
    merged_list = merged_list.next
print()

print("---------------")

# Example 2
# Creating linked lists: a: 1->1, b: 2->4
a1 = Node(1)
a2 = Node(1)
a1.next = a2
list_a = a1

b1 = Node(2)
b2 = Node(4)
b1.next = b2
list_b = b1


merged_list = merge_sorted_lists(list_a, list_b)

while merged_list:
    print(merged_list.data, end=" ")
    merged_list = merged_list.next
print()
```

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.prev = None
        self.next = None

def reverse_doubly_linked_list(head):
    if not head:
        return None

    current = head
    prev = None

    while current:
        temp = current.prev
        current.prev = current.next
        current.next = temp
        prev = current
        current = current.prev

    head = prev

    return head

# Example
# Creating a doubly linked list: 10->8->4->2
n1 = Node(10)
n2 = Node(8)
n3 = Node(4)
n4 = Node(2)
n1.next = n2
n2.prev = n1
n2.next = n3
n3.prev = n2
n3.next = n4
n4.prev = n3
linked_list = n1

reversed_list = reverse_doubly_linked_list(linked_list)

while reversed_list:
    print(reversed_list.data, end=" ")
    reversed_list = reversed_list.next
print()
```

```python
class Node:
    def __init__(self, data=None):
        self.data = data
        self.prev = None
        self.next = None

def delete_node_at_position(head, position):
    if not head:
        return head

    if position == 1:
        next_node = head.next
        if next_node:
            next_node.prev = None
        head = next_node
        return head

    count = 1
    current = head

    while current and count < position:
        current = current.next
        count += 1

    if current.next:
        current.prev.next = current.next
        current.next.prev = current.prev
    else:
        
        current.prev.next = None

    return head

# Example 1
# Creating a doubly linked list: 1<->3<->4
n1 = Node(1)
n2 = Node(3)
n3 = Node(4)
n1.next = n2
n2.prev = n1
n2.next = n3
n3.prev = n2
linked_list = n1

new_list = delete_node_at_position(linked_list, 3)

while new_list:
    print(new_list.data, end=" ")
    new_list = new_list.next
print()

print("------------")
# Example 2
# Creating a doubly linked list: 1<->5<->2<->9
n1 = Node(1)
n2 = Node(5)
n3 = Node(2)
n4 = Node(9)
n1.next = n2
n2.prev = n1
n2.next = n3
n3.prev = n2
n3.next = n4
n4.prev = n3
linked_list = n1

new_list = delete_node_at_position(linked_list, 1)

while new_list:
    print(new_list.data, end=" ")
    new_list = new_list.next
print()
```
